### What is Big-O?
Big O notation is a fundamental concept in computer science and data science that allows us to analyze the efficiency and performance of algorithms. It provides a standardized way of expressing the upper bound of an algorithm's time complexity in relation to the input size. In simpler terms, it helps us understand how the execution time of an algorithm grows as the input size increases.

We will navigate through various chapters, each focusing on a specific aspect of Big O notation. From the simplicity of constant time complexity (O(1)) to the complexities of exponential time (O(2^n)) and factorial time (O(n!)), our journey will encompass a range of algorithms and scenarios commonly encountered in data science.

### 1. O(1) - Constant Time Complexity
The notation O(1) represents constant time complexity, where the execution time remains constant regardless of the input size. This is often associated with simple operations or data access, appending to a list or removal. 

It is the most efficient algorithms in terms of time complexity.

```python
def constant_example(array):
    return array[0]
```

In the above example, accessing the first element of the data structure takes constant time, irrespective of the size of the data.

##### Tip: To avoid mistakes, ensure that the operation truly takes constant time and is not dependent on the input size.

### 2. O(n) - Linear Time Complexity
Linear time complexity, denoted as O(n), implies that the execution time of an algorithm grows linearly with the size of the input. Common operations with linear time complexity include iterating through a list or array, inserting to a list (as we dont know at what position it will be inserted), lookup / search in array, or Window sliding algorithm (https://www.geeksforgeeks.org/window-sliding-technique/) or Monotonic Stack (https://www.geeksforgeeks.org/introduction-to-monotonic-stack-data-structure-and-algorithm-tutorials/)

```python
def linear_example(array):
    for item in array:
        print(item)
```

Here, the time it takes to iterate through the entire data is directly proportional to the size of the data.

##### Tip: Avoid inefficiencies by optimizing loops and minimizing unnecessary iterations.

### 3. O(n^2) - Quadratic Time Complexity
Quadratic time complexity, O(n^2), indicates that the execution time grows quadratically with the input size. Nested loops are a common source of quadratic time complexity.

```python
def quadratic_example(array):
    for i in array:
        for j in array:
            print(i, j)
```

In this example, the nested loops result in a quadratic growth in execution time as the input size increases.


But many times we will only have to go through half of the combinations (ie. just [i,j], but not [j,i] combination). This is a n^2/2 complexity, still O(n^2), but half the time. For such case we can use the code bellow:

```python
def every_pair_quadratic(array):
    for i in range(len(array)):
        for j in range(i+1, len(array)):
            print(array[i], array[j])
```

Insertion sort is also O(n^2) complexity.
##### Tip: To enhance efficiency of sorting, explore alternative sorting algorithms like QuickSort or MergeSort.

### 4. O(n * m) - Polynomial Time Complexity
Polynomial time complexity with O(n * m) signifies an algorithm's performance that depends on two input sizes. This often occurs with nested loops involving different input sizes.

```python
def polynomial_example(array1, array2):
    for i in array1:
        for j in array2:
            print(i, j)
```

The time complexity here depends on both the size of array1 and array2.

##### Tip: Optimize algorithms with polynomial time complexity by identifying opportunities for parallelization or using more efficient data structures.

### 5. O(n^3) - Cubic Time Complexity
Cubic time complexity, O(n^3), indicates that the execution time grows cubically with the input size. This usually involves three nested loops, or three-dimensional spaces. Usually this is the highest of power that is used in the algorithms.

```python
def cubic_example(array):
    for i in array:
        for j in array:
            for k in array:
                print(i, j, k)
```

As seen in this example, the nested loops result in cubic growth in execution time.

##### Tip: Efficient algorithms with cubic time complexity can often be optimized by rethinking the problem or exploring parallel computing.

### 6. O(logn) - Logarithmic Time Complexity
Logarithmic time complexity, O(logn), is characteristic of algorithms that divide the problem in half at each step. A classic example would be the Binary search (hot or cold game, or left/right tree search). Over a big input size the O(logn) becomes basically flat and much more efficient than O(n).

```python
def binary_search(array, target):
    low, high = 0, len(array) - 1
    while low <= high:
        mid = (low + high) // 2
        if array[mid] == target:
            return mid
        elif array[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

In binary search, the size of the problem is reduced by half in each iteration, resulting in a logarithmic growth in execution time.

##### Tip: Minimize the search space efficiently to optimize logarithmic time complexity algorithms.

### 7. O(nlogn) - Linearithmic Time Complexity
Linearithmic time complexity, O(nlogn), is often associated with efficient sorting algorithms like MergeSort and HeapSort. Similarly to comparison between O(logn) and O(n), O(nlogn) is much more effecient than O(n^2).

```python
def merge_sort(array):
    if len(array) <= 1:
        return array

    mid = len(array) // 2
    left = merge_sort(array[:mid])
    right = merge_sort(array[mid:])

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result
```

In merge sort, the array is repeatedly divided into halves, and the merging process is linear. This results in a total time complexity of O(nlogn).

##### Tip: Optimize sorting algorithms and ensure efficient merging techniques to maintain O(nlogn) complexity.

## Non-polynomial algorithms
### 8. O(2^n) - Exponential Time Complexity
Exponential time complexity, O(2^n), indicates an algorithm where the execution time doubles with each additional element in the input. It is most common in the recursion. 

```python
def fibonacci_recursive(n):
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)
```

The classic example of calculating Fibonacci numbers using recursion results in exponential growth in execution time.

##### Tip: Mitigate the impact of exponential time complexity by implementing dynamic programming or memoization techniques.

### 9. O(sqrt(n)) - Square Root Time Complexity

### 10. O(n!) - Factorial Time Complexity
Factorial time complexity, O(n!), signifies an algorithm whose execution time grows factorially with the input size. This is rear and highly inefficient, and is typically associated with brute force approaches. In comes up in Permutations and Graph problems like Travelling Salesman Problem (https://www.geeksforgeeks.org/traveling-salesman-problem-tsp-implementation/).


```python
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n-1)
```

Calculating the factorial of a number using recursion is an example of an algorithm with factorial time complexity.

##### Tip: Avoid factorial time complexity by redesigning algorithms or identifying more efficient solutions.

### Bonus: Let's plot these algorithms

```python
import matplotlib.pyplot as plt
import numpy as np

# Input sizes
input_sizes = np.arange(1, 13)  # Adjust the range based on your needs

# Functions representing different complexity classes
def constant_time(input_sizes):
    return np.ones_like(input_sizes)

def linear_time(input_sizes):
    return input_sizes

def quadratic_time(input_sizes):
    return input_sizes ** 2

def polynomial_time(input_sizes):
    return input_sizes * np.log(input_sizes)  # Example of polynomial time, adjust as needed

def cubic_time(input_sizes):
    return input_sizes ** 3

def logarithmic_time(input_sizes):
    return np.log2(input_sizes)

def nlogn_time(input_sizes):
    return input_sizes * np.log2(input_sizes)

def exponential_time(input_sizes):
    return 2 ** input_sizes

def sqrt_time(input_sizes):
    return np.sqrt(input_sizes)

def factorial_time(input_sizes):
    return [np.math.factorial(n) for n in input_sizes]

# Plotting
plt.figure(figsize=(12, 6))

plt.plot(input_sizes, constant_time(input_sizes), label="O(1) - Constant Time")
plt.plot(input_sizes, linear_time(input_sizes), label="O(n) - Linear Time")
plt.plot(input_sizes, quadratic_time(input_sizes), label="O(n^2) - Quadratic Time")
plt.plot(input_sizes, polynomial_time(input_sizes), label="O(n * m) - Polynomial Time")
plt.plot(input_sizes, cubic_time(input_sizes), label="O(n^3) - Cubic Time")
plt.plot(input_sizes, logarithmic_time(input_sizes), label="O(log(n)) - Logarithmic Time")
plt.plot(input_sizes, nlogn_time(input_sizes), label="O(n * log(n)) - Linearithmic Time")
plt.plot(input_sizes, exponential_time(input_sizes), label="O(2^n) - Exponential Time")
plt.plot(input_sizes, sqrt_time(input_sizes), label="O(sqrt(n)) - Square Root Time")
plt.plot(input_sizes, factorial_time(input_sizes), label="O(n!) - Factorial Time")

# Adding data labels
for func, label in zip([constant_time, linear_time, quadratic_time, polynomial_time, cubic_time,
                        logarithmic_time, nlogn_time, exponential_time, sqrt_time, factorial_time],
                       ["O(1)", "O(n)", "O(n^2)", "O(n * m)", "O(n^3)", "O(log(n))",
                        "O(n * log(n))", "O(2^n)", "O(sqrt(n))", "O(n!)"]):
    plt.text(input_sizes[-1] + 0.2, func(input_sizes)[-1], label, fontsize=8)
    
plt.yscale("log")  # Log scale for better visibility of different classes
plt.xlabel("Input Size")
plt.ylabel("Operations")
plt.title("Big O Notation Complexity Classes")
plt.legend()
plt.grid(True)
plt.show()
```
